```python
import torch
import torch.nn as nn
class CustomClassifier(nn.Module):
    def __init__(self, in_features, num_classes=2):
        super(CustomClassifier, self).__init__()
        
        self.fc1_1=nn.Linear(in_features=in_features, out_features=num_classes)
        

    def forward(self, x):
        
        x = self.fc1_1(x)
       
        return x
model = torch.load(r'C:\Users\Vansh\Desktop\projects\Maverick\Wild fire\wf_densenet169.pth')
model.eval()
```

```python
from PIL import Image
import torchvision.transforms as transforms

# Load the test image
image_path =r"C:\Users\Vansh\Desktop\projects\Maverick\deforestation\a1.png"
image = Image.open(image_path).convert('RGB')

# Define the image transformations (resize, normalize, etc.)
preprocess = transforms.Compose([
    transforms.Resize((128,128)),  # Resize the image to the size expected by the model
    transforms.ToTensor(),          # Convert the image to a PyTorch tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize the image
])

# Preprocess the image
input_tensor = preprocess(image)
input_batch = input_tensor.unsqueeze(0)

```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
input_batch = input_batch.to(device)

# Ensure the model is in evaluation mode
model.eval()

# Run the model on the test image
with torch.no_grad():
    output = model(input_batch)
```

```python
probabilities = torch.nn.functional.softmax(output[0], dim=0)

# Get the class with the highest probability
_, predicted_class = torch.max(probabilities, 0)

# If you have a list of class names, map the predicted class index to the class name
class_names = ['nowildfire','wildfire']  # Replace with your actual class names
predicted_label = class_names[predicted_class]

print(f'Predicted label: {predicted_label}')
```
